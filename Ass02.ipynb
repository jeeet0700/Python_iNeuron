11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

```python
for i in range(10):
    print(i)
"""It will print the entired value from default(0) value to <nth value(10)"""

for i in range(0,10):
    print(i)
"""It will print the value from 0 to <10th, the diffrent is, in the above we are defining the start and stop value
lower bound and upper bound"""

for i in range(0,10,1):
    print(i)
"""It will print the value from 0 to 10 with a jump/step size of 1"""
```
