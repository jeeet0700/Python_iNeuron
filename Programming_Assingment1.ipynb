Write a Python program to swap two variables?

```python
a,b=10,20
temp=0
print("Before swap A : {}".format(a))
print("Before swap B : {}".format(b))
temp=b
b=a
a=temp
print("After swap A : {}".format(a))
print("After swap B : {}".format(b))
```
